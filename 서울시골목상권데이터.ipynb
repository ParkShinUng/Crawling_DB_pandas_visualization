```python
import requests
from bs4 import BeautifulSoup
import re
import json

#인구수
url = 'https://golmok.seoul.go.kr/region/selectPopulation.json'

req_param = {
    'stdrYyCd': 2020,
    'stdrQuCd': 2,
    'svcIndutyCdL': 'CS000000',
    'svcIndutyCdM': 'all'
}
# 요청을 보내기
res = requests.post(url, data=req_param)
print(type(res))
# 응답코드 200(ok) 가 정상이면
if res.ok:
    #response 객체에 포함된 text를 추출
    html = res.text
    #BeautifulSoup 객체의 역할 - html 문서 파싱
    #텍스트를 추출(parsing)하기 위해서 응답 text와 parser 이름을 전달한다.
    soup = BeautifulSoup(html, 'lxml')
    
    print(type(soup))
#     print(soup)
    data = soup.select('p')[0].text
    print(type(data))
    json_data = json.loads(data)
    print(type(json_data))
#    print(json_data)
#     for row in json_data:
#         print(type(row))
#         print('-----')
```

```python
import pandas as pd

data_df = pd.DataFrame(json_data)
print(data_df.shape)
print(data_df.columns)
data_df.head()
```
